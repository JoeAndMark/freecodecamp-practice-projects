```python
import pandas as pd
```

```python
df = pd.read_csv("epa-sea-level.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.scatter(x=df["Year"], y=df["CSIRO Adjusted Sea Level"])
```

```python
from scipy.stats import linregress
```

```python
res = linregress(x=df["Year"], y=df["CSIRO Adjusted Sea Level"])
```

```python
res
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.scatter(x=df["Year"], y=df["CSIRO Adjusted Sea Level"])
years_extended = pd.Series(range(1880, 2051))
plt.plot(years_extended, res.intercept + res.slope*years_extended, 'r', scalex=True)
```

```python
df_limited = df[df["Year"] >= 2000]
res = linregress(x=df_limited["Year"], y=df_limited["CSIRO Adjusted Sea Level"])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
```

```python
plt.scatter(x=df["Year"], y=df["CSIRO Adjusted Sea Level"])
years_extended = pd.Series(range(1880, 2051))
plt.plot(years_extended, res.intercept + res.slope*years_extended, 'r', scalex=True)
```
